```python
data =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)
ans =data.shape
ans
```

```python
data
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}
data['new_income']=data['Income_Category'].map(dic)
ans = data['new_income']
ans.head()
```

```python
data['AgeState'] = data['Customer_Age'].map(lambda x : x//10*10)
ans = data['AgeState'].value_counts().sort_index()
ans
```

```python
data.loc[:,'newEduLevel']=data['Education_Level'].map(lambda x: 1 if 'Graduate' in x else 0)
data['newEduLevel'].value_counts()
```

```python
data.loc[:,'newLimit']=data['Credit_Limit'].map(lambda x: 1 if x>=4500 else 0)
data['newLimit'].value_counts()
```

```python
def chg(x):
    if x['Marital_Status']=='Married' and x['Card_Category']=='Platinum':
        return 1
    else:
        return 0
data.loc[:,'newState']=data[['Marital_Status','Card_Category']].apply(chg,axis=1)
data['newState'].value_counts()
```

```python
def g_chg(x):
    if x['Gender']=='M':
        return 'male'
    else:
        return 'female'
data.loc[:,'Gender']=data[['Gender']].apply(g_chg,axis=1)
data['Gender'].value_counts()
```

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
data = pd.read_csv(DataUrl)
data.head()
```

```python
import datetime
def data_ft(x):
    if x[-2:]!='24':
        x = x.replace(':', ' ')
        return pd.to_datetime(x)
    else:
        date = pd.to_datetime(x.split(':')[0])+datetime.timedelta(days=1)
        return date
data['(년-월-일:시)'] = data['(년-월-일:시)'].apply(data_ft)
```

```python
data['dayName']=data['(년-월-일:시)'].dt.day_name()
data['dayName'].head(6)
```
