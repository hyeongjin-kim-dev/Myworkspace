# 머신러닝이란?

Def) 데이터에서부터 학습하도록 컴퓨터를 프로그래밍하는 과학

- 훈련세트 : 시스템이 학습하는 데 사용하는 샘플
- 훈련사례(샘플) : 훈련 데이터
- 정확도 : 성능 측정이며, 분류 작업제 자주 사용된다.

# 전통적인 접근 vs 머신러닝 기법

#### 전통적인 접근

- 문제가 어렵기 때문에 규칙이 점점 길고 복잡해지므로 유지 보수하기 매우 힘들다.

1. 예시
    1. 먼저 스팸에 어떤 단어들이 주로 나타나는지 살펴본다. 그러면 특정 단어가 구절이나 제목에 많이 나타나는 경향을 볼 수 있다. 이뿐 아니라 보낸이의 이름, 매일 주소, 본문 등 다른 요소에서도 패턴을 감지 할 수 있다.
    1. 발견한 각 패턴을 감지하는 알고리즘을 작성하여 프로그램이 이런 패턴을 발견했을 때 그 매일을 스팸으로 분류
    1. 프로그램을 테스트하고 론칭할 만큼 충분한 성능이 나올 때까지 1~2단계 반복

#### 머신러닝 기법

- 프로그램이 훨씬 짧아지고 유지 보수하기 쉬우며 정확도가 높다.

1. 예시
    1. 스팸에 자주 나타나는 패턴을 감지하여 어떤 단어와 구절이 스팸 메일을 판단하는 데 좋은 기준인지 자동으로 학습

2. 비교
    1. 전통적인 프로그래밍 방식에서는 바뀔 때 마다 수정을 해줘야한다 영원히 새로운 규칙을 추가해야한다.
    1. 머신러닝 기반은 바뀔 때마다 자동으로 인식하고 별도의 작업을 진행하지 않아도 자동으로 스팸으로 분류

##### 그림 예시

>머신러닝을 통해 우리가 학습을 배울 수도 있습니다. 알고리즘이 학습한 것을 조사 할 수 있다.
>>- 예시) 스팸 필터가 충분한 수팸 메일로 훈련되었따면 스팸을 예측하는 데 가장 좋은 단어 및 단어의 조합이
무엇인지 확인 할 수 있다.

>- 데이터마이닝: 머신러닝 기술을 적용해서 대용량의 데이터를 분석하면
    __겉으로는 보이지 않던 패턴을 발견 하는 것__

>

# 머신러닝 시스템의 종류

- 사람의 감독하에 훈련하는 것인지 그렇지 않은 것인지(지도, 비지도, 준지도, 강화학습)
- 실시간으로 점진적인 학습을 하는지 아닌(온라인 학습과 배치학습)
- 단순하게 알고 있는 데이터 포인트와 새 데이터 포인트를 비교하는 것인지 아니면 과학자들이 하는 것처럼 훈련 데이터 셋에서 패턴을 발견하여 예측 모델을 만드는지(사례 기반 학습과 모델 기반 학습)

#### 지도 학습 vs 비지도 학습 vs 준지도 vs 강화

>지도학습

+ 지도학습에는 알고리즘에 주입하는 훈련 데이터에 __레이블__이라는 답이 포함
    + 분류 : 전형적인 지도 학습이다.
    + 회귀 : 예측 변수(특성[Feature])을 사용해 타깃(Target)수치를 예측하는 것

+ 지도 학습 알고리즘
    + K-최근접 이웃(KNN)
    + 선형 회귀(Linear Regression)
    + 로지스틱 회귀(Logistic Regression)
    + 서포트 벡터 머신(Support Vector Machine)[SVM]
    + 결정 트리(Decision tree) & 랜덤 포레스트(Random Forest)
    + 신경망(Neural networks)

> 비지도 학습 -훈련 데이터에 레이브링 없다.

+ 비지도 학습 알고리즘
    + 군집
        + K-평균(K-means)
        + DBSCAN
        + 계층 군집 분석(Hierarchical cluster analysis)[HCA]
        + 이상치 탐지(Outlier Detection) & 특이치 탐지(Novelty Detection)
        + 원-클래스(One-class SVM)
        + 아이솔레이션 포레스트(Isolation Forest)
    + 시각화와 차원 축소
        + 주성분 분석(Principal component analysis)(PCA)
        + 커널(y-Kernel)PCA
        + 지역적 선형 임배딩(Locally - linear - embeadding)
        + t-SNE
    + 연관 규칙 학습
        + 어프라이어리
        + 이클렛

+ 예시
    + 이상치 탐지 - 훈련하는 동안 대부분 정상 샘플을 만나 이를 인식하도록 훈련하며
     새로운 샘플을 보고 정상 데이터인지 혹은 이상치인지 판단하는 것
    + 특이치 탐지 - 훈련 세트에 있는 모든 샘플과 달라 보이는 새로운 샘프을 탐지하는
    것이 목적
    + 연관 규칙 학습 - 대량의 데이터에서 특성 간의 흥미로운 관계츷 찾는것

> 준지도 학습 - 일부만 레이블이 있는 데이터

+ 준지도 학습 알고리즘 - 지도학습 + 비지도 학습
    + 심층 신뢰 신경망(Deep Belief Network)[DBN] -여러 겹으로 쌓은 제한된 볼츠만 머신(RBM)에 기초

> 강화학습 - 여기서는 학습하는 시스템을 에이전트라 부르며 환경을 관찰해서 행동을 실행하고 그 결과로 보상, 부정적인 보상은 벌점

# 배치학습 & 온라인 학습

> 배치학습

오프라인 학습 - 시스템을 훈련시키고 그런 다음 제품 시스템에 적용하면 더 이상의 학습없이 실행되며, 학습한 것을 단지 적용만 하는 것

> 온라인 학습 - 데이터를 순차적으로 한 개씩 또는 미니배치라 부르는 작은 묶은 단위로 주입하여 시스템을 훈련시킨다.

# 사례 기반 학습과 모델 기반 학습

> 사례 기반 학습 - 시스템이 훈련샘플을 기억하고 학습한 뒤, 유사도 측정을 사용하여 새로운 데이터와 학습한 샘플을 비교하여 일반화

> 모델 기반 학습 - 샘플모델을 만들고 예측에 사용하는 것

1. 모델이 최상의 성능을 내도록 하는 값을 알 수 있는 법..?
    1. 효용함수 : 모델이 얼마나 좋은지 측정
    1. 비용함수 : 얼마나 나쁜지 측정

# 머신러닝의 주요 도전 과제

1. 충분하지 않은 양의 훈련 데이터
    + 수천개 수만개의 데이터가 필요하며, 이미지나 음성 인식 같은 복잡한 데이터는 수백만게가 필요할 수도 있다.


2. 대표성 없는 훈련 데이터
   + 일반화하려는 사례들을 대표하는 훈련세트를 사용하는 것이 매우 중요하나 어렵다. 샘플이 작으면 __샘플링 잡음__ , 매우 큰 샘플도 표본 추출 방법이 잘못되면 대표성을 못 띈는데 이를 __샘플링 편향__ 이라고 한다

3. 낮은 품질의 데이터
    + 훈련데이터가 정제 되지 않은 상태에서 머신러닝 시스템은 내재된 패턴을 찾기 어려워 잘 작동하지 않는다.
    

    3-1 낮은 품질의 데이터가 정제가 필요한 경우
       - 일부 샘플이 이상치라는 게 확실하면 간단히 무시하거나 수동으로 잘못된 것으로 고쳐야 좋다.
       - 일부 샘플에 특성 몇 개가 빠져있다면, 그 특성을 어떻게(버릴지, 다른 값 대체, 다른 모델과 비교)할지 결정해야한다.
            

4. 관련없는 특성 - 좋은 특성을 찾아야 머신러닝 프로젝트는 성공
    + 특성 선택 : 가지고 있는 특성 중에서 훈련에 가장 유용한 특성을 선택한다.
    + 특성 추출 : 특성을 결합하여 더 유용한 특성을 만든다. 앞서 본 것처럼 차원 축소 알고리즘이 도움이 된다.
    + 새로운 데이터를 수집해 새 특성을 만든다.

5. 과대적합 - 과도한 일반화를 뜻함, 모델이 훈련 데이터에는 너무 잘 맞지만 일반성이 떨어진다.
    + 규제 : 모델을 단순하게 하고 과대적합의 위험을 감소시키기 위해 모델에 제약을 가하는 것
    + 자유도 : 모델은 직선의 절편과 기울기를 조절 할 수 있다. 한쪽을 0이 되도록 강제하면 알고리즘 상의 하나의 자유도로 데이터에 적절하게 맞추는 것은 어렵다.
    + 하이퍼파라미터 : 학습 알고리즘 상의 파라미터(인위적)

6. 과소적합 - 모델이 너무 단순해서 데이터의 내재된 구조를 학습하지 못해서 발생
    1. 해결법
        + 모델 파라미터가 더 많은 강력한 모델을 선택한다.
        + 학습 알고리즘에 더 좋은 특성을 제공합니다.(특성 공학)
        + 모델의 제약을 줄인다.(ex. 규제 하이퍼파라미터를 감소)

# 테스트와 검증

1. 훈련세트 & 테스트 세트로 나누기
    + 훈련 세트를 사용해 모델을 훈련한 뒤 테스트 세트에 사용을 통해 모델을 테스트합니다.
    + 일반화 오차 : 새로운 샘플에 대한 오류 비율
    + 테스트 세트에서 얻은 추정값을 통해 새로운 샘플에 모델이 얼마나 잘 작동하는지 확인

2. 하이퍼파라미터 튜닝과 모델 선택
    + 홀드아웃 검증 : 훈련 세트의 일부를 떼어내어 여러 후보 모델을 평가하고 가장 좋은 하나를 선택  ---> 검증 세트
    + 줄어든 훈련세트에서 다양한 하이퍼파라미터 값을 가진 모델 훈련 -> 검증 세트에서 성능이 가장 좋은 모델 선택 -> 전체 훈련세트에 다시 훈련 -> 최종모델 데트스 세트에서 평가 후 일반화 오차 추정
    + 교차검증 : 검증세트마다 나머지 데이터에서 훈련한 모델을 해당 검증 세트에서 평가
